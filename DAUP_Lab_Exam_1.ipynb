```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
numerical_feature = ['area']
bin_sizes = [10, 20]

for feature, bin_size in zip(numerical_feature, bin_sizes):
    plt.figure(figsize=(10, 20))
    sns.histplot(data=data, x=feature, kde=True, bins=bin_size)
    plt.show()
```

```python
numerical_feature = ['price']
bin_sizes = [10, 20]

for feature, bin_size in zip(numerical_feature, bin_sizes):
    plt.figure(figsize=(10, 20))
    sns.histplot(data=data, x=feature, kde=True, bins=bin_size)
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='area', data=data)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='price', data=data)
plt.show()
```
